# Market Basket Analysis in Python
Apriori Rule Mining in python

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

#### Load Data

```python
df = pd.read_csv('Data/Final.csv')
```

#### Show Unclean Data
Shows an instance of complete dataset. Note that some values are negative as there was an error while conversion from pdf to csv. Cleaning is done in later cells.

```python
df.head()
```

#### Cleaning of Data

```python
# Clean up spaces in description and remove any rows that don't have a valid invoice
#df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['Inv.#'], inplace=True)

df['Inv.#'] = df['Inv.#'].astype('str')
df = df[~df['Inv.#'].str.contains('C')]
```

#### Convert Data to Appropriate Format Used in Associate Rule Mining

```python
basket = (df
          .groupby(['Inv...', 'Product.Name'])['Qty']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Inv...'))
```

```python
basket
```

```python
# Show a subset of columns
basket.iloc[:,[0,1,2,3,4,5,6, 7]].head()
```

#### One Hot Encoded

```python
# Convert the units to 1 hot encoded values
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1    
```

```python
basket_sets = basket.applymap(encode_units)
```

### Data after Cleaning 

```python
basket_sets
```

### Association Rule Mining by Aprior

```python
# Build up the frequent itemsets with apriori
frequent_itemsets = apriori(basket_sets, min_support=0.2, use_colnames=True)
```

```python
frequent_itemsets
```

```python
# Create the rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# Filter the rules
rules[ (rules['lift'] >= 2) &
       (rules['confidence'] >= 0.1) ]
```

```python
# Total times atenolol tab occoured in dataset
basket['ATENOLOL TAB'].sum()
```

### Apriori with Different Minimum Support

```python
frequent_itemsets2 = apriori(basket, min_support=0.05, use_colnames=True)
```

#### Show the Discovered Rules
Note that rules have increased a lot. 

```python
frequent_itemsets2
```

```python
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1)
rules2
```

```python
# Fitler Rules 
rules2[ (rules2['lift'] >= 2.5) &
        (rules2['confidence'] >= 0.1) ]
```
